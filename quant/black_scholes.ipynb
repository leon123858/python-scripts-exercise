Black-Scholes æ¨¡å‹å°‡é‡‘èå¸‚å ´çš„ä¸ç¢ºå®šæ€§é‡åŒ–ç‚ºæ•¸å­¸å…¬å¼ã€‚é›–ç„¶å®ƒæœ‰å…¶å±€é™æ€§ï¼Œä½†å®ƒæä¾›äº†ä¸€å€‹æ¨™æº–åŒ–çš„èªè¨€ï¼Œè®“äº¤æ˜“å“¡èƒ½å¤ æ¯”è¼ƒåƒ¹æ ¼ä¸¦ç®¡ç†é¢¨éšªã€‚

## åŸºç¤å…¬å¼æ‡‰ç”¨

ä»¥ä¸‹ç›´æ¥å¯¦ä½œæ¨¡å‹çš„å…¬å¼ï¼Œå¸¶å…¥æŒ‡å®šåƒæ•¸

- K é¸æ“‡æ¬Šè¦æ±‚æˆäº¤æ™‚æ‰€éœ€æ”¯ä»˜çš„åƒ¹æ ¼
- T è·é›¢å±¥ç´„æ™‚é–“å–®ä½
- r æ¯æ™‚é–“å–®ä½çš„ç„¡é¢¨éšªåˆ©ç‡
- sigma æ¯æ™‚é–“å–®ä½æ³¢å‹•ç‡

note: æ™‚é–“å–®ä½é•·åº¦éœ€è¦çµ±ä¸€

- ç•¶ä¸‹é¸æ“‡æ¬Šåˆç†çš„åƒ¹æ ¼

concept

- å–®ä½æŠ•è³‡é‡‘é¡ * ç„¡é¢¨éšªåˆ©ç‡ = æ¯å–®ä½è³‡é‡‘çš„ç„¡é¢¨éšªç²åˆ©
- æ¯å–®ä½è³‡é‡‘çš„ç„¡é¢¨éšªç²åˆ© = æ¯å–®ä½è³‡é‡‘é¸æ“‡æ¬Šçš„æœŸæœ›ç²åˆ© - æ¯å–®ä½è³‡é‡‘é¸æ“‡æ¬Šçš„åˆç†åƒ¹æ ¼
- æ¯å–®ä½è³‡é‡‘é¸æ“‡æ¬Šçš„åˆç†åƒ¹æ ¼ = æ¯å–®ä½è³‡é‡‘é¸æ“‡æ¬Šçš„æœŸæœ›ç²åˆ© - æ¯å–®ä½è³‡é‡‘çš„ç„¡é¢¨éšªç²åˆ©
- è²·æ¬Šåƒ¹æ ¼ = mean(å¯èƒ½çš„æœªä¾†è‚¡åƒ¹ - å±¥ç´„åƒ¹) - ç¾é‡‘æˆæœ¬

```python
import numpy as np
from scipy.stats import norm

def black_scholes_call(S, K, T, r, sigma):
    """
    è¨ˆç®—æ­å¼è²·æ¬Š (Call Option) çš„ Black-Scholes åƒ¹æ ¼

    åƒæ•¸:
    S     : æ¨™çš„è³‡ç”¢ç¾åƒ¹ (Spot Price)
    K     : å±¥ç´„åƒ¹ (Strike Price)
    T     : è·é›¢åˆ°æœŸæ™‚é–“ (Time to Maturity, å–®ä½: å¹´)
    r     : ç„¡é¢¨éšªåˆ©ç‡ (Risk-free Rate, ä¾‹å¦‚ 0.05 ä»£è¡¨ 5%)
    sigma : æ³¢å‹•ç‡ (Volatility, ä¾‹å¦‚ 0.2 ä»£è¡¨ 20%)

    å›å‚³:
    call_price : é¸æ“‡æ¬Šç†è«–åƒ¹æ ¼
    """

    # è™•ç†é‚Šç•Œæƒ…æ³ï¼šå¦‚æœ T è¶¨è¿‘æ–¼ 0ï¼Œç›´æ¥å›å‚³ Payoff
    if T <= 1e-6:
        return max(0, S - K)

    # 1. è¨ˆç®— d1 å’Œ d2
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    # 2. è¨ˆç®— N(d1) å’Œ N(d2) -> é€™è£¡å°æ‡‰å…¬å¼ä¸­çš„æ©Ÿç‡é …
    N_d1 = norm.cdf(d1)
    N_d2 = norm.cdf(d2)

    # 3. å¸¶å…¥å…¬å¼: Asset Term - Liability Term
    # C = S * N(d1) - K * e^(-rT) * N(d2)
    call_price = S * N_d1 - K * np.exp(-r * T) * N_d2

    return call_price

# ==========================================
# ğŸ§ª å¯¦é©—å®¤ï¼šè«‹åˆ©ç”¨é€™è£¡çš„åƒæ•¸é€²è¡Œæ¸¬è©¦
# ==========================================

# å®šç¾©ä¸€å€‹è¼”åŠ©å‡½å¼ä¾†æ¼‚äº®åœ°åˆ—å°çµæœ
def run_scenario(name, S, K, T, r, sigma):
    price = black_scholes_call(S, K, T, r, sigma)
    print(f"--- {name} ---")
    print(f"åƒæ•¸: S={S}, K={K}, T={T}, r={r}, sigma={sigma}")
    print(f"ğŸ’° é¸æ“‡æ¬Šç†è«–åƒ¹æ ¼: {price:.4f}")
    print("-" * 30)

# é¡Œç›® 1: åŸºæº–æ¸¬è©¦ (Baseline)
# æƒ…å¢ƒï¼šè‚¡åƒ¹ 100ï¼Œå±¥ç´„åƒ¹ 100 (åƒ¹å¹³ ATM)ï¼Œé‚„æœ‰ä¸€å¹´åˆ°æœŸï¼Œæ³¢å‹•ç‡ 20%ï¼Œåˆ©ç‡ 5%
run_scenario("é¡Œç›® 1: åŸºæº–æ¸¬è©¦", S=100, K=100, T=1.0, r=0.05, sigma=0.2)

# é¡Œç›® 2: é«˜é¢¨éšªé«˜å ±é…¬ (High Volatility)
# æƒ…å¢ƒï¼šè·Ÿé¡Œç›® 1 ä¸€æ¨£ï¼Œä½†é€™å®¶å…¬å¸åƒåŠ å¯†è²¨å¹£ä¸€æ¨£æ³¢å‹•åŠ‡çƒˆ (æ³¢å‹•ç‡è®Š 50%)
# é æ¸¬ï¼šåƒ¹æ ¼æ‡‰è©²æœƒè®Šè²´é‚„æ˜¯è®Šä¾¿å®œï¼Ÿ
run_scenario("é¡Œç›® 2: é«˜æ³¢å‹•ç‡", S=100, K=100, T=1.0, r=0.05, sigma=0.5)

# é¡Œç›® 3: æ™‚é–“ä¸å¤šäº† (Time Decay)
# æƒ…å¢ƒï¼šè·Ÿé¡Œç›® 1 ä¸€æ¨£ï¼Œä½†æ˜å¤©å°±è¦åˆ°æœŸäº† (T = 1/365 ç´„ 0.0027 å¹´)
# é æ¸¬ï¼šæ—¢ç„¶æ˜¯åƒ¹å¹³ (ATM)ï¼Œæ™‚é–“åƒ¹å€¼æ‡‰è©²æ‰€å‰©ç„¡å¹¾ï¼Ÿ
run_scenario("é¡Œç›® 3: è‡¨è¿‘åˆ°æœŸ", S=100, K=100, T=0.0027, r=0.05, sigma=0.2)

# é¡Œç›® 4: æ·±åº¦åƒ¹å…§ (Deep ITM) - æ¥µé™æ¸¬è©¦
# æƒ…å¢ƒï¼šè‚¡åƒ¹ 1000ï¼Œå±¥ç´„åƒ¹ 10 (å¹¾ä¹ä¿è­‰å±¥ç´„)ï¼Œæ³¢å‹•ç‡å’Œåˆ©ç‡ä¸è®Š
# é æ¸¬ï¼šåƒ¹æ ¼æ‡‰è©²éå¸¸æ¥è¿‘ S - K * e^(-rT)
run_scenario("é¡Œç›® 4: æ·±åº¦åƒ¹å…§", S=1000, K=10, T=1.0, r=0.05, sigma=0.2)
```

æ¨å°çš„æ ¸å¿ƒé‚è¼¯æ˜¯ï¼š**é€éã€Œè²·å…¥é¸æ“‡æ¬Šã€ä¸¦ã€Œæ”¾ç©ºè‚¡ç¥¨ã€ï¼Œæ§‹å»ºä¸€å€‹é¢¨éšªå®Œå…¨æŠµéŠ·çš„æŠ•è³‡çµ„åˆï¼Œä¸¦ä»¤å…¶æ”¶ç›Šç­‰æ–¼ç„¡é¢¨éšªåˆ©ç‡ã€‚**

## ç¬¬ä¸€æ­¥ï¼šæè¿°è³‡ç”¢åƒ¹æ ¼è¡Œç‚º (å¹¾ä½•å¸ƒæœ—é‹å‹•)

é¦–å…ˆï¼Œå‡è¨­è‚¡åƒ¹ $S$ éµå¾ªå¹¾ä½•å¸ƒæœ—é‹å‹• (Geometric Brownian Motion, GBM)ï¼š

$$
dS = \mu S dt + \sigma S dW
$$

* $\mu$ï¼šè‚¡ç¥¨é æœŸå ±é…¬ç‡ (Drift)
* $\sigma$ï¼šæ³¢å‹•ç‡ (Volatility)
* $dW$ï¼šéŸ‹ç´éç¨‹ (Wiener Process)ï¼Œä»£è¡¨éš¨æ©Ÿé …ï¼Œæ»¿è¶³ $dW \sim N(0, dt)$

## ç¬¬äºŒæ­¥ï¼šæè¿°é¸æ“‡æ¬Šåƒ¹æ ¼è¡Œç‚º (ItÃ´'s Lemma)

å‡è¨­é¸æ“‡æ¬Šåƒ¹æ ¼ $V$ æ˜¯è‚¡åƒ¹ $S$ å’Œæ™‚é–“ $t$ çš„å‡½æ•¸ï¼Œå³ $V(S, t)$ã€‚
æ ¹æ“š **ItÃ´'s Lemma (ä¼Šè—¤å¼•ç†)**ï¼Œå‡½æ•¸ $V$ çš„å¾®åˆ† $dV$ å¯ä»¥å±•é–‹ç‚ºæ³°å‹’ç´šæ•¸ (ä¿ç•™åˆ°äºŒéšé …)ï¼š

$$
dV = \frac{\partial V}{\partial t} dt + \frac{\partial V}{\partial S} dS + \frac{1}{2} \frac{\partial^2 V}{\partial S^2} (dS)^2
$$


note æ³°å‹’å±•é–‹

$$
dV = \frac{\partial V}{\partial t} dt + \frac{\partial V}{\partial S} dS + \frac{1}{2} \frac{\partial^2 V}{\partial S^2} (dS)^2 + \frac{1}{2} \frac{\partial^2 V}{\partial t^2} (dt)^2 + \frac{\partial^2 V}{\partial S \partial t} dS dt + \dots
$$

æˆ‘å€‘çŸ¥é“ $(dS)^2$ çš„å„é …è¦å‰‡ç‚ºï¼š
* $dt \cdot dt \approx 0$
* $dt \cdot dW \approx 0$
* $(dW)^2 = dt$

æ‰€ä»¥ï¼Œ$(dS)^2 = (\sigma S dW)^2 = \sigma^2 S^2 dt$ã€‚

å°‡ $dS$ å’Œ $(dS)^2$ ä»£å› $dV$ çš„å¼å­ä¸­ï¼š

$$
dV = \left( \frac{\partial V}{\partial t} + \mu S \frac{\partial V}{\partial S} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} \right) dt + \sigma S \frac{\partial V}{\partial S} dW
$$

é€™å€‹å¼å­å‘Šè¨´æˆ‘å€‘é¸æ“‡æ¬Šåƒ¹æ ¼è®Šå‹•åŒ…å«å…©å€‹éƒ¨åˆ†ï¼š
1.  **ç¢ºå®šæ€§éƒ¨åˆ† ($dt$é …)**ï¼šæ™‚é–“æµé€å’Œè‚¡åƒ¹è¶¨å‹¢çš„å½±éŸ¿ã€‚
2.  **éš¨æ©Ÿé¢¨éšªéƒ¨åˆ† ($dW$é …)**ï¼šè‚¡åƒ¹æ³¢å‹•å¸¶ä¾†çš„é¢¨éšªã€‚

## ç¬¬ä¸‰æ­¥ï¼šå»ºæ§‹ç„¡é¢¨éšªæŠ•è³‡çµ„åˆ ($\Pi$)

ç‚ºäº†æ¶ˆé™¤é¢¨éšª (ä¹Ÿå°±æ˜¯æ¶ˆé™¤ $dW$ é …)ï¼Œæˆ‘å€‘å»ºæ§‹ä¸€å€‹æŠ•è³‡çµ„åˆ $\Pi$ï¼š
* è²·å…¥ 1 å–®ä½é¸æ“‡æ¬Š ($V$)
* è³£ç©º (Short) $\Delta$ å–®ä½çš„è‚¡ç¥¨ ($S$)

$$
\Pi = V - \Delta S
$$

è©²çµ„åˆåœ¨æ¥µçŸ­æ™‚é–“å…§çš„åƒ¹å€¼è®Šå‹• $d\Pi$ ç‚ºï¼š

$$
d\Pi = dV - \Delta dS
$$

å°‡å‰é¢æ¨å°çš„ $dV$ å’Œ $dS$ ä»£å…¥ï¼š

$$
d\Pi = \underbrace{\left[ \left( \frac{\partial V}{\partial t} + \mu S \frac{\partial V}{\partial S} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} \right) dt + \sigma S \frac{\partial V}{\partial S} dW \right]}_{dV} - \Delta \underbrace{(\mu S dt + \sigma S dW)}_{dS}
$$

æ¥è‘—ï¼Œæˆ‘å€‘æŠŠå«æœ‰ $dW$ (é¢¨éšªé …) å’Œå«æœ‰ $dt$ (ç¢ºå®šé …) çš„éƒ¨åˆ†æ•´ç†åœ¨ä¸€èµ·ï¼š

$$
d\Pi = \left( \frac{\partial V}{\partial t} + \mu S \frac{\partial V}{\partial S} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} - \Delta \mu S \right) dt + \left( \sigma S \frac{\partial V}{\partial S} - \Delta \sigma S \right) dW
$$

## ç¬¬å››æ­¥ï¼šDelta å°æ²– (æ¶ˆé™¤é¢¨éšª)

è¦è®“é€™å€‹æŠ•è³‡çµ„åˆè®Šæˆã€Œç„¡é¢¨éšªã€ï¼Œå¿…é ˆè®“éš¨æ©Ÿé … $dW$ çš„ä¿‚æ•¸ç‚ºé›¶ã€‚ä¹Ÿå°±æ˜¯ï¼š

$$
\sigma S \frac{\partial V}{\partial S} - \Delta \sigma S = 0
$$

ç”±æ­¤å¾—å‡ºè‘—åçš„ **Delta å°æ²–æ¯”ç‡**ï¼š

$$
\Delta = \frac{\partial V}{\partial S}
$$

é€™æ„å‘³è‘—å¦‚æœä½ æŒæœ‰çš„ç©ºé ­è‚¡ç¥¨æ•¸é‡æ­£å¥½ç­‰æ–¼é¸æ“‡æ¬Šå°è‚¡åƒ¹çš„æ•æ„Ÿåº¦ ($\frac{\partial V}{\partial S}$)ï¼Œé¢¨éšªå°±è¢«å®Œå…¨å°æ²–äº†ã€‚æ­¤æ™‚ï¼Œ$d\Pi$ ä¸­åªå‰©ä¸‹ $dt$ é …ï¼š

$$
d\Pi = \left( \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} \right) dt
$$
*(æ³¨æ„ï¼šåŸæœ¬çš„ $\mu S$ é …å› ç‚º $\Delta = \frac{\partial V}{\partial S}$ äº’ç›¸æŠµéŠ·äº†ï¼Œé€™è­‰æ˜äº†é¸æ“‡æ¬Šå®šåƒ¹èˆ‡è‚¡ç¥¨é æœŸå ±é…¬ç‡ $\mu$ ç„¡é—œã€‚)*

## ç¬¬äº”æ­¥ï¼šæ‡‰ç”¨ç„¡å¥—åˆ©æ¢ä»¶ ($d\Pi = r \Pi dt$)

é€™å°±æ˜¯ä½ æŒ‡å®šçš„èµ·é»ã€‚æ—¢ç„¶ $\Pi$ ç¾åœ¨æ˜¯ç„¡é¢¨éšªçš„ï¼Œæ ¹æ“š**å–®ä¸€åƒ¹æ ¼æ³•å‰‡**ï¼Œå®ƒçš„æ”¶ç›Šç‡å¿…é ˆç­‰æ–¼ç„¡é¢¨éšªåˆ©ç‡ $r$ã€‚

$$
d\Pi = r \Pi dt
$$

æˆ‘å€‘å°‡å·¦é‚Šçš„ $d\Pi$ (æ­¥é©Ÿå››çš„çµæœ) å’Œå³é‚Šçš„ $\Pi$ (æ­¥é©Ÿä¸‰çš„å®šç¾©ï¼Œä¸¦ä»£å…¥ $\Delta = \frac{\partial V}{\partial S}$) ä»£å…¥ç­‰å¼ï¼š

$$
\underbrace{\left( \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} \right) dt}_{d\Pi} = r \underbrace{\left( V - S \frac{\partial V}{\partial S} \right)}_{\Pi} dt
$$

æ¶ˆå»å…©é‚Šçš„ $dt$ï¼Œå¾—åˆ°ï¼š

$$
\frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} = rV - rS \frac{\partial V}{\partial S}
$$

## ç¬¬å…­æ­¥ï¼šBlack-Scholes åå¾®åˆ†æ–¹ç¨‹ (PDE)

å°‡ä¸Šè¿°ç­‰å¼ç§»é …æ•´ç†ï¼Œè®“æ‰€æœ‰é …åœ¨ä¸€é‚Šï¼Œç­‰æ–¼ 0ï¼š

$$
\frac{\partial V}{\partial t} + rS \frac{\partial V}{\partial S} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} - rV = 0
$$

é€™å°±æ˜¯è‘—åçš„ **Black-Scholes PDE**ã€‚ä»»ä½•æ­å¼è¡ç”Ÿæ€§å•†å“ï¼ˆè²·æ¬Šã€è³£æ¬Šï¼‰çš„åƒ¹æ ¼ $V$ éƒ½å¿…é ˆæ»¿è¶³é€™å€‹æ–¹ç¨‹ã€‚

---

## ç¬¬ä¸ƒæ­¥ï¼šæ±‚è§£ PDE (å¾—åˆ°æœ€çµ‚å…¬å¼)

è¦å¾é€™å€‹å¾®åˆ†æ–¹ç¨‹å¾—åˆ°æˆ‘å€‘å¸¸è¦‹çš„ $N(d_1)$ å…¬å¼ï¼Œéœ€è¦è§£é€™å€‹ PDEã€‚é€™é€šå¸¸éœ€è¦åˆ©ç”¨é‚Šç•Œæ¢ä»¶ (Boundary Conditions) ä¸¦é€éè®Šæ•¸è®Šæ›å°‡å…¶è½‰åŒ–ç‚ºç‰©ç†å­¸ä¸Šçš„ **ç†±å‚³å°æ–¹ç¨‹å¼ (Heat Equation)** ä¾†æ±‚è§£ã€‚

**å°æ–¼æ­å¼è²·æ¬Š (Call Option)ï¼Œé‚Šç•Œæ¢ä»¶ç‚ºï¼š**
1.  **åˆ°æœŸæ™‚åƒ¹å€¼**ï¼š$V(S, T) = \max(S - K, 0)$
2.  **è‚¡åƒ¹ç‚º 0 æ™‚**ï¼š$V(0, t) = 0$
3.  **è‚¡åƒ¹ç„¡é™å¤§æ™‚**ï¼š$V(S, t) \approx S$

**æ±‚è§£ç°¡è¿°ï¼š**
é€éè®Šæ•¸è®Šæ›ï¼š
Let $x = \ln(S)$, $\tau = T-t$, $v(x, \tau) = e^{-r\tau}V(S,t)$ ...ç­‰ä¸€ç³»åˆ—æ•¸å­¸æ“ä½œï¼Œå¯ä»¥å°‡ BS PDE è½‰æ›ç‚ºæ¨™æº–ç†±å‚³å°æ–¹ç¨‹ï¼š

$$
\frac{\partial u}{\partial \tau} = \frac{\partial^2 u}{\partial x^2}
$$

åˆ©ç”¨ç†±å‚³å°æ–¹ç¨‹çš„ç©åˆ†é€šè§£ï¼Œä»£å…¥ä¸Šè¿°é‚Šç•Œæ¢ä»¶é€²è¡Œç©åˆ†é‹ç®—ï¼Œæœ€çµ‚å°±æœƒæ¨å°å‡ºï¼š

$$
C(S, t) = S_0 N(d_1) - K e^{-rT} N(d_2)
$$

å…¶ä¸­ï¼š
$$
d_1 = \frac{\ln(S_0/K) + (r + \sigma^2/2)T}{\sigma\sqrt{T}}
$$
$$
d_2 = d_1 - \sigma\sqrt{T}
$$

é€™å°±æ˜¯å¾ç„¡å¥—åˆ©æ¢ä»¶ ($d\Pi = r\Pi dt$) ä¸€è·¯æ¨å°è‡³æœ€çµ‚å…¬å¼çš„å®Œæ•´é‚è¼¯è·¯å¾‘ã€‚

```python
import sympy as sp

# åˆå§‹åŒ–å°å‡ºæ¼‚äº®çš„æ•¸å­¸æ ¼å¼
sp.init_printing(use_unicode=True, wrap_line=False, use_latex='mathjax')

# 1. å®šç¾©ç¬¦è™Ÿ
# mu: é æœŸå ±é…¬ç‡, sigma: æ³¢å‹•ç‡, t: æ™‚é–“, r: ç„¡é¢¨éšªåˆ©ç‡ (å¾ŒçºŒæœƒç”¨åˆ°)
mu, sigma, t, r = sp.symbols('mu sigma t r')

# dt: æ™‚é–“å¢é‡, dW: ç¶­ç´éç¨‹å¢é‡
dt, dW = sp.symbols('dt dW')

# S: è‚¡åƒ¹ï¼Œå®ƒæ˜¯æ™‚é–“ t çš„å‡½æ•¸ (ä½†åœ¨ SDE ä¸­é€šå¸¸ç°¡å¯«ç‚º S)
S = sp.symbols('S')

# 2. å®šç¾© GBM æ–¹ç¨‹å¼
# è‚¡åƒ¹è®ŠåŒ– = é æœŸè®ŠåŒ–*è‚¡åƒ¹*æ™‚é–“è®ŠåŒ– + æ³¢å‹•ç‡*è‚¡åƒ¹*æ³¢å‹•è®ŠåŒ–
dS = mu * S * dt + sigma * S * dW

fix_S_delta = mu * S * dt
random_S_delta = sigma * S * dW

# 3. ç°¡åŒ–æ–¹ç¨‹å¼
# è‚¡åƒ¹è®ŠåŒ– = å›ºå®šè®ŠåŒ– + éš¨æ©Ÿæ³¢å‹•
# dS = fix_S_delta + random_S_delta

# print("--- Step 1: å®šç¾© GBM æ¨¡å‹ ---")
# print("è³‡ç”¢åƒ¹æ ¼è®Šå‹• dS å®šç¾©ç‚ºï¼š")
display(sp.Eq(sp.Symbol('dS'), dS))
# display_latex_image(fr"{sp.latex(dS)}")
```

```python
import sympy as sp

# V æ˜¯é¸æ“‡æ¬Šåƒ¹æ ¼
# S è‚¡åƒ¹
# t æ™‚é–“ (å±¥ç´„åˆ°æœŸæ™‚é–“, ä¸å¯ææ—©å±¥ç´„)
# å®šç¾© V ç‚º S å’Œ t çš„æœªå®šå‡½æ•¸
V = sp.Function('V')(S, t)

# 1. å¯«å‡ºæ³°å‹’å±•é–‹å¼ (è¶…éäºŒéšéƒ¨åˆ†å¿½ç•¥)
# print("--- 1. æ³°å‹’ç´šæ•¸å±•é–‹ (Taylor Expansion) ---")
dV_taylor = sp.Derivative(V, t) * dt + sp.Derivative(V, S) * dS + sp.Derivative(V, S, S) * (dS**2) * 0.5 + sp.Derivative(V, t, t) * (dt**2) * 0.5
# print("dV (raw) =")
# sp.pprint(dV_taylor)
# print("\n")

# 2. æ‡‰ç”¨ Ito Rules åŒ–ç°¡
ito_rules = {
    dW**2: dt,      # å¸ƒæœ—é‹å‹•å‡è¨­: æ¥µå°æ³¢å‹• = K * dt^(1/2) , æ‰€ä»¥ æ¥µå°æ³¢å‹• = K * dt^2
    dt**2: 0,       # æ™‚é–“å–æ¥µçŸ­ï¼Œæ¥µçŸ­æ™‚é–“å¹³æ–¹ç‚º 0
    dt*dW: 0,       # æ¥µå°æ™‚é–“çš„æ³¢å‹•è¦–ç‚º 0
    dW*dt: 0        # ä¸Šå¼äº¤æ›ç‡
}

dV_fully_expanded = sp.expand(dV_taylor)
dV_ito_simplified = dV_fully_expanded.subs(ito_rules)

# print("dV (simple) =")
# sp.pprint(dV_taylor)
# print("\n")

# 3. æ•´ç†çµæœ (æŒ‰ dt å’Œ dW åˆ†çµ„)
dV_final = sp.collect(dV_ito_simplified, [dt, dW])

# print("--- 3. æ‡‰ç”¨ Ito Rules åŒ–ç°¡å¾Œçš„ dV ---")
# print("å°‡ (dW)^2 æ›¿æ›ç‚º dtï¼Œä¸¦å¿½ç•¥é«˜éšé …å¾Œï¼š")
display(sp.Eq(sp.Symbol('dV'), dV_final))
# display_latex_image(fr"{sp.latex(dV_final)}")
# æœ‰ dW*sigma çš„ part åŒ…å«æ³¢å‹•æ˜¯é¢¨éšªéƒ¨ä½ï¼Œç´”ç²¹åŸºæ–¼ S*mu / S*sigma æ˜¯é è¨­çš„æˆé•·ç‡å’Œè®ŠåŒ–ç‡ï¼Œæ˜¯å›ºå®šéƒ¨ä½
```

```python
import sympy as sp
from IPython.display import display

# ---------------------------------------------------------
# æ¨å°æŠ•è³‡çµ„åˆçš„åƒ¹å€¼è®Šå‹• (Derive dPi)
# ---------------------------------------------------------
# æ¥è‘—è¦æ¨å°ä¸€å€‹æŠ•è³‡çµ„åˆçš„å…¬å¼:
# æŠ•è³‡çµ„åˆåƒ¹å€¼(Pi) = åšå¤šéƒ¨ä½ - è³£ç©ºéƒ¨ä½
#                = è²·é€²ä¸€å–®ä½é¸æ“‡æ¬Šåƒ¹æ ¼(V) - è³£ç©º n å–®ä½è‚¡åƒ¹(S)
#                = V - n * S
#
# æˆ‘å€‘è¨ˆç®—å…¶å¾®å°è®Šå‹•é‡: dPi = dV - n * dS
#
# ç›®æ¨™:
# 1. å°‡ä¸Šé¢æ¨å°å‡ºçš„ dV (Ito Lemmaçµæœ) èˆ‡ dS (GBMå®šç¾©) å¸¶å…¥å…¬å¼ã€‚
# 2. ç›¡åŠ›åŒ–ç°¡æ•¸å­¸å¼ã€‚
# 3. å°‡çµæœä¾æ“š dt (ç¢ºå®šæ€§é …) èˆ‡ dW (éš¨æ©Ÿé¢¨éšªé …) é€²è¡Œåˆ†çµ„æ•´ç†ã€‚
# ---------------------------------------------------------

# å®šç¾© n: æˆ‘å€‘è³£ç©ºè‚¡ç¥¨çš„å–®ä½æ•¸é‡ (é€™è£¡ç”¨ n è¡¨ç¤º)
n = sp.symbols('n')
Pi = V - n*S
# ç›´æ¥æ‰‹ç®—ï¼Œå·¦å³åŒå¾®åˆ†
# 1. è¨ˆç®— dPi = dV - n * dS
dPi = dV_final - n * dS

# 2. å±•é–‹ä¸¦ä¾æ“š dt å’Œ dW é€²è¡Œåˆ†çµ„æ•´ç† (Collect)
# é€™æ˜¯æœ€é‡è¦çš„ä¸€æ­¥ï¼Œæˆ‘å€‘å« SymPy æŠŠæ‰€æœ‰è·Ÿ dt æœ‰é—œçš„æ”¾åœ¨ä¸€èµ·ï¼Œè·Ÿ dW æœ‰é—œçš„æ”¾åœ¨ä¸€èµ·
dPi_expanded = sp.expand(dPi)
dPi_grouped = sp.collect(dPi_expanded, [dt, dW])

# 3. ç‚ºäº†è®“é¡¯ç¤ºæ›´æ¼‚äº®ï¼Œæˆ‘å€‘æŠŠ dW çš„ä¿‚æ•¸å†æ¬¡æå–å…¬å› å¼ (ä¾‹å¦‚æå– sigma * S)
# é›–ç„¶ dPi_grouped å·²ç¶“æ­£ç¢ºï¼Œä½†æ‰‹å‹•æ•´ç†ä¸€ä¸‹é¡¯ç¤ºæœƒæ›´ç›´è¦º
coeff_dW = dPi_grouped.coeff(dW)
coeff_dt = dPi_grouped.coeff(dt)

# é‡æ–°çµ„åˆä¸€å€‹ã€Œè¦–è¦ºä¸Šã€æ¼‚äº®çš„ç­‰å¼
dPi_visual = sp.Eq(sp.Symbol('Pi'), coeff_dt * dt + coeff_dW * dW)
# print("--- Step 3: æŠ•è³‡çµ„åˆåƒ¹å€¼è®Šå‹• dPi ---")
# print("å·²å°‡ dV å’Œ dS ä»£å…¥ï¼Œä¸¦æŒ‰ dt èˆ‡ dW åˆ†çµ„æ•´ç†ï¼š")
display(dPi_visual)
```

```python
# ---------------------------------------------------------
# Step 4: Delta å°æ²– (Delta Hedging)
# ---------------------------------------------------------
# ç‚ºäº†è®“æŠ•è³‡çµ„åˆæˆç‚ºç„¡é¢¨éšªè³‡ç”¢ (Risk-Free)ï¼Œ
# æˆ‘å€‘å¿…é ˆæ¶ˆé™¤ dPi ä¸­çš„éš¨æ©Ÿé … (dW)ã€‚
#
# é€éèª¿æ•´ n å¯ä»¥æ¶ˆå»é¢¨éšªé … dWï¼Œå› æ­¤å¯ä»¥å¾—åˆ°ç†è«–ä¸Šçš„ç„¡é¢¨éšªå…¬å¼
# å¯¦å‹™ä¸Šåªè¦åˆ©ç”¨è©² n çš„æ¯”ä¾‹æ“ä½œé¸æ“‡æ¬Šï¼Œå¯ä»¥ç†è«–å°æ²–æ‰ä»»ä½•é¢¨éšª
# è¦è®“ coeff_dW = 0
#
# æ“ä½œæ­¥é©Ÿ:
# 1. å–å‡º dW çš„ä¿‚æ•¸ (coeff_dW)ã€‚
# 2. ä»¤ coeff_dW = 0ï¼Œä¸¦æ±‚è§£ n (è³£ç©ºæ¯”ä¾‹)ã€‚
# 3. å°‡æ±‚å¾—çš„ n ä»£å›åŸæœ¬çš„ dPi å…¬å¼ï¼Œå¾—åˆ°ç„¡é¢¨éšªçš„ dPi (åªå‰©ä¸‹ dt é …)ã€‚
# ---------------------------------------------------------

# 1. å–å‡º dW çš„ä¿‚æ•¸
# dPi_grouped æ˜¯ä¸Šä¸€æ­¥æ•´ç†å¥½çš„è®Šæ•¸
coeff_dW = dPi_grouped.coeff(dW)

# print("--- 1. dW çš„ä¿‚æ•¸æ–¹ç¨‹å¼ ---")
# æˆ‘å€‘å»ºç«‹ä¸€å€‹ç­‰å¼: ä¿‚æ•¸ = 0
equation_risk_neutral = sp.Eq(coeff_dW, 0)
# display(equation_risk_neutral)

# 2. æ±‚è§£ n
# sp.solve æœƒå›å‚³ä¸€å€‹åˆ—è¡¨ï¼Œæˆ‘å€‘å–ç¬¬ä¸€å€‹è§£
n_solution = sp.solve(equation_risk_neutral, n)[0]

# print("--- 2. æ±‚å¾—ç„¡é¢¨éšªè³£ç©ºæ¯”ä¾‹ n (å³ Delta) ---")
display(sp.Eq(n, n_solution))

# 3. å°‡è§£ä»£å› dPi
# é€™æ¨£å°±å¾—åˆ°äº†ã€Œç„¡é¢¨éšªæŠ•è³‡çµ„åˆã€çš„åƒ¹å€¼è®Šå‹•å…¬å¼
dPi_riskless = dPi_grouped.subs(n, n_solution)

# å†æ¬¡æ•´ç† (é›–ç„¶ç†è«–ä¸Š dW æ²’äº†ï¼Œä½†ç‚ºäº†ä¹¾æ·¨ï¼Œæˆ‘å€‘å†åšä¸€æ¬¡ collect dt)
dPi_riskless = sp.collect(sp.expand(dPi_riskless), dt)

# print("--- 3. ä»£å…¥ n å¾Œçš„ç„¡é¢¨éšª dPi ---")
# print("æ³¨æ„è§€å¯Ÿï¼šdW é …æ¶ˆå¤±äº†ï¼Œä¸” mu (é æœŸå ±é…¬ç‡) ä¹Ÿäº’ç›¸æŠµéŠ·æ¶ˆå¤±äº†ï¼")
display(sp.Eq(sp.Symbol("dPI"), dPi_riskless))
```

```python
# ---------------------------------------------------------
# Step 5: æ‡‰ç”¨ç„¡å¥—åˆ©æ¢ä»¶å¾—å‡º PDE (Apply No-Arbitrage Condition)
# ---------------------------------------------------------
# æ ¹æ“šç„¡å¥—åˆ©åŸç†:
# ä¸€å€‹ç„¡é¢¨éšªçš„æŠ•è³‡çµ„åˆ PI ï¼Œå…¶å ±é…¬ç‡å¿…é ˆç­‰æ–¼å¸‚å ´ç„¡é¢¨éšªåˆ©ç‡ rã€‚
# å³: dPi = r * Pi * dt
#
# æ“ä½œæ­¥é©Ÿ:
# 1. å»ºç«‹ç­‰å¼: dPi_riskless = r * Pi * dt
# 2. åŒ–ç°¡ç­‰å¼ (å…©é‚Šæ¶ˆå» dt)ã€‚
# 3. ç§»é …æ•´ç†ï¼Œå¾—åˆ°æ¨™æº–çš„ Black-Scholes PDE (= 0 çš„å½¢å¼)ã€‚
# ---------------------------------------------------------

# print("--- 1. å»ºç«‹ç„¡å¥—åˆ©æ–¹ç¨‹å¼ (dPi = r * Pi * dt) ---")
raw_pi = Pi.subs(n, n_solution)
arbitrage_eq = sp.Eq(dPi_riskless, r * raw_pi * dt)
display(arbitrage_eq)

# 2. åŒ–ç°¡èˆ‡ç§»é …
# å…©é‚ŠåŒæ™‚é™¤ä»¥ dt (å› ç‚º dt ä¸ç‚º 0)
# æ³¨æ„: åœ¨ SymPy ä¸­ï¼Œæˆ‘å€‘å¯ä»¥ç›´æ¥å°‡ç­‰å¼å®šç¾©ç‚º (lhs - rhs = 0) ä¸¦é™¤ä»¥ dt
pde_raw = (dPi_riskless - r * raw_pi * dt) / dt

# å±•é–‹ä¸¦æ•´ç†
pde_final = sp.expand(pde_raw)

# ç‚ºäº†ç¬¦åˆæ•™ç§‘æ›¸æ¨™æº–æ ¼å¼ (V_t + ... - rV = 0)ï¼Œæˆ‘å€‘æŒ‰å°æ•¸éšæ•¸æ•´ç†
# collect åªæ˜¯ç‚ºäº†é¡¯ç¤ºæ¼‚äº®ï¼Œæ•¸å­¸ä¸Šæ˜¯ä¸€æ¨£çš„
pde_sorted = sp.collect(pde_final, [sp.Derivative(V, t), sp.Derivative(V, S), sp.Derivative(V, S, S), V])

# print("--- 2. æœ€çµ‚æ¨å°å‡ºçš„ Black-Scholes PDE ---")
# print("ç§»é …æ•´ç†ä½¿å¾—ç­‰å¼ç­‰æ–¼ 0ï¼š")
# å»ºç«‹ä¸€å€‹æ¼‚äº®çš„ç­‰å¼é¡¯ç¤º
final_pde_display = sp.Eq(pde_sorted, 0)
display(final_pde_display)
```

```python
# æœ€å¾Œ step æ˜¯åšè®Šæ•¸è®Šæ›èˆ‡è§£ PDE
# æ±‚å¾— V
# æ­¤æ­¥è·³éï¼Œç›´æ¥æ¼”ç¤ºè¨ˆç®—é©—è­‰

import sympy as sp

# 1. å®šç¾©ç¬¦è™Ÿ
S0, K, r, sigma, T = sp.symbols('S_0 K r sigma T', real=True, positive=True)
# S_T: åˆ°æœŸæ™‚çš„è‚¡åƒ¹è®Šæ•¸
ST = sp.symbols('S_T', real=True, positive=True)

# 2. å®šç¾©åˆ°æœŸè‚¡åƒ¹ S_T çš„åˆ†ä½ˆ
# æ ¹æ“šå¹¾ä½•å¸ƒæœ—é‹å‹•ï¼Œln(S_T) æœå¾å¸¸æ…‹åˆ†ä½ˆ:
# ln(S_T) ~ N(ln(S0) + (r - 0.5*sigma^2)*T, sigma^2 * T)

mu_log = sp.log(S0) + (r - sp.Rational(1, 2) * sigma**2) * T
sigma_log = sigma * sp.sqrt(T)

# å»ºç«‹ä¸€å€‹æ¨™æº–å¸¸æ…‹åˆ†ä½ˆè®Šæ•¸ Z (ç‚ºäº†æ–¹ä¾¿ç©åˆ†æ¨å°)
x = sp.symbols('x', real=True) # ç©åˆ†è®Šæ•¸

# å°æ•¸å¸¸æ…‹åˆ†ä½ˆçš„æ©Ÿç‡å¯†åº¦å‡½æ•¸ (PDF)
# é€™è£¡æˆ‘å€‘æ‰‹å‹•å¯«å‡º PDF ä»¥ç¢ºä¿ SymPy èƒ½è™•ç†
# PDF formula for LogNormal: 1/(x*s*sqrt(2pi)) * exp(...)
pdf = 1 / (x * sigma_log * sp.sqrt(2 * sp.pi)) * sp.exp(
    -((sp.log(x) - mu_log)**2) / (2 * sigma_log**2)
)
# æˆ‘å€‘ç›´æ¥å‘Šè¨´ SymPy é€™æ˜¯æˆ‘å€‘æƒ³è¦çš„çµæœå½¢å¼ (ç”¨æ¨™æº–å¸¸æ…‹ç´¯ç©åˆ†å¸ƒå‡½æ•¸ CDF è¡¨ç¤º)
# å› ç‚º SymPy ç©åˆ†å‡ºçš„åŸå§‹çµæœæœƒåŒ…å« erf (Error Function)ï¼Œä¸ç›´è§€
# æˆ‘å€‘å®šç¾©æ¨™æº–å¸¸æ…‹åˆ†ä½ˆçš„ CDF å‡½æ•¸ N(x)
def N(z):
    return sp.Rational(1, 2) * (1 + sp.erf(z / sp.sqrt(2)))

# å®šç¾© d1 å’Œ d2
d1 = (sp.log(S0 / K) + (r + sp.Rational(1, 2) * sigma**2) * T) / (sigma * sp.sqrt(T))
d2 = d1 - sigma * sp.sqrt(T)

# æ¨™æº–çš„ Black-Scholes å…¬å¼
BS_closed_form = S0 * N(d1) - K * sp.exp(-r * T) * N(d2)

print("ç‚ºäº†è­‰æ˜çµæœï¼Œæˆ‘å€‘å±•ç¤ºæ¨™æº– Black-Scholes å…¬å¼å½¢å¼ï¼š")
display(sp.Eq(sp.Symbol('V'), BS_closed_form))

print("\n--- é©—è­‰ç©åˆ†çµæœ (Verification) ---")
print("æˆ‘å€‘ç”¨ä¸€çµ„æ•¸å­—å¸¶å…¥é©—è­‰ï¼šç©åˆ†çµæœ vs å…¬å¼çµæœ")

# è¨­å®šä¸€çµ„æ¸¬è©¦æ•¸å­—
values = {S0: 100, K: 100, T: 1, r: 0.05, sigma: 0.2}

# 2. è¨ˆç®—å…¬å¼æ•¸å€¼ (Closed Form)
val_formula = BS_closed_form.evalf(subs=values)

print(f"é€é BS å…¬å¼è¨ˆç®—:      {val_formula}")
```

- å°æ¯å€‹æ¨£æœ¬ä½œ GBM å…¬å¼è§£
- é‡å°æ¯å€‹æ¨£æœ¬çµæœç®—ç²åˆ©
- è—‰ç”±ç„¡é¢¨éšªåˆ©ç‡ä¿®æ­£ç²åˆ©
- æœ€çµ‚ç²åˆ©å°±æ˜¯é¸æ“‡æ¬Šçš„åƒ¹å€¼


```python
import numpy as np
from scipy.stats import norm
import time

# 1. ç‚ºäº†é©—è­‰ï¼Œæˆ‘å€‘å…ˆæ”¾ä¸Šå‰›å‰›å¯«å¥½çš„ BS å…¬å¼å‡½æ•¸
def black_scholes_analytic(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# 2. è’™åœ°å¡ç¾…æ¨¡æ“¬æ¼”ç®—æ³• (Monte Carlo Simulation)
def monte_carlo_option_pricing(S, K, T, r, sigma, num_simulations=1000000):
    """
    S: ç¾åƒ¹
    K: å±¥ç´„åƒ¹
    T: åˆ°æœŸæ™‚é–“
    r: ç„¡é¢¨éšªåˆ©ç‡ (æ³¨æ„ï¼šåœ¨é¢¨éšªä¸­æ€§å®šåƒ¹ä¸­ï¼Œæˆ‘å€‘ç”¨ r å–ä»£æ¼‚ç§»ç‡ mu)
    sigma: æ³¢å‹•ç‡
    num_simulations: æ¨¡æ“¬æ¬¡æ•¸ (æ¨£æœ¬æ•¸)
    """
    np.random.seed(42) # å›ºå®šç¨®å­ä»¥ä¾¿é‡ç¾çµæœ

    # --- æ­¥é©Ÿ A: ç”¢ç”Ÿéš¨æ©Ÿè®Šæ•¸ ---
    # æ ¹æ“š GBM çš„è§£: S_T = S_0 * exp( (r - 0.5*sigma^2)T + sigma*sqrt(T)*Z )
    # Z æ˜¯æ¨™æº–å¸¸æ…‹åˆ†ä½ˆéš¨æ©Ÿè®Šæ•¸ N(0, 1)
    # é•·åº¦ç‚º num_simulations çš„éš¨æ©Ÿæ•¸å­—é™£åˆ—ï¼Œéµå¾ªæ¨™æº–å¸¸æ…‹åˆ†ä½ˆ
    z = np.random.standard_normal(num_simulations)

    # --- æ­¥é©Ÿ B: è¨ˆç®—åˆ°æœŸæ—¥çš„è‚¡åƒ¹ (S_T) ---
    # é€™è£¡ç›´æ¥ä½¿ç”¨å‘é‡åŒ–é‹ç®—ï¼Œä¸€æ¬¡ç®—å‡ºä¸€ç™¾è¬å€‹ S_T
    # é€™æ˜¯ å¹¾ä½•å¸ƒæœ—é‹å‹• (GBM) çš„ç©åˆ†å½¢å¼è§£ï¼ŒST æ˜¯é•·åº¦ç‚º num_simulations çš„ä¸åŒ z ä¹‹é‹ç®—çµæœ
    ST = S * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * z)
    # print(len(z), len(ST))

    # --- æ­¥é©Ÿ C: è¨ˆç®— Payoff ---
    # Payoff = max(ST - K, 0)
    # payoffs æ˜¯è¨ˆç®—æ¯å€‹é‹ç®—çµæœçš„ç²åˆ©é‡‘é¡
    payoffs = np.maximum(ST - K, 0)

    # --- æ­¥é©Ÿ D: æŠ˜ç¾å›ä»Šå¤© ---
    # é¸æ“‡æ¬Šåƒ¹æ ¼ = å¹³å‡ Payoff * æŠ˜ç¾å› å­
    # æŠ˜ç¾å› å­ çš„å…¬å¼ä¾†è‡ª é€£çºŒè¤‡åˆ© (Continuous Compounding) çš„é€†é‹ç®—
    # æŠ˜ç¾å› å­è¦è€ƒæ…®ç„¡é¢¨éšªåˆ©ç‡å’Œæ™‚é–“ï¼Œå¯ä»¥ç¨±å¾—ä¸Šæ˜¯: ç„¡é¢¨éšªå¯ä»¥ç²åˆ©çš„é‡‘é¡æ¯”ä¾‹
    option_price = np.mean(payoffs) * np.exp(-r * T)

    # è¨ˆç®—æ¨™æº–èª¤ (Standard Error) ä»¥è©•ä¼°ç²¾ç¢ºåº¦
    std_error = np.std(payoffs) / np.sqrt(num_simulations)

    return option_price, std_error

# ==========================================
# ğŸš€ å¯¦æˆ°æ¸¬è©¦
# ==========================================

# åƒæ•¸è¨­å®š
S_0 = 100      # ç¾åƒ¹
K = 100        # å±¥ç´„åƒ¹
T = 1.0        # 1å¹´
r = 0.05       # åˆ©ç‡ 5%
sigma = 0.2    # æ³¢å‹•ç‡ 20%
sims = 100000000 # æ¨¡æ“¬

print(f"--- é–‹å§‹è¨ˆç®— (æ¨¡æ“¬æ¬¡æ•¸: {sims:,}) ---")

# 1. è¨ˆç®—è§£æè§£ (æ¨™æº–ç­”æ¡ˆ)
start_time = time.time()
bs_price = black_scholes_analytic(S_0, K, T, r, sigma)
print(f"ğŸ“˜ Black-Scholes è§£æè§£: {bs_price:.4f}")

# 2. è¨ˆç®—è’™åœ°å¡ç¾…æ¨¡æ“¬è§£ (æ•¸å€¼è¿‘ä¼¼)
mc_price, mc_error = monte_carlo_option_pricing(S_0, K, T, r, sigma, sims)
end_time = time.time()

print(f"ğŸ² Monte Carlo æ¨¡æ“¬çµæœ:  {mc_price:.4f}")
print(f"   (æ¨™æº–èª¤: +/- {mc_error:.4f})")
print(f"â±ï¸ æ¨¡æ“¬è€—æ™‚: {end_time - start_time:.4f} ç§’")

# 3. æ¯”è¼ƒèª¤å·®
diff = abs(mc_price - bs_price)
print("-" * 30)
print(f"ğŸ” èª¤å·® (Difference): {diff:.4f}")
print(f"   èª¤å·®ç™¾åˆ†æ¯”: {diff/bs_price*100:.4f}%")
```
